````python
# BCCWJ 頻度データを取得
from termcolor import colored
from zipfile import ZipFile
import requests
import os

# BCCWJ の短単位と長単位のデータを定義
bccwj_info = {
    'short': {'url':'https://repository.ninjal.ac.jp/?action=repository_uri&item_id=3234&file_id=22&file_no=1',
              'zip_fname': 'BCCWJ_frequencylist_suw_ver1_0.zip',
              'freq': []},
    'long': {'url': 'https://repository.ninjal.ac.jp/?action=repository_uri&item_id=3228&file_id=22&file_no=1',
             'zip_fname': 'BCCWJ_frequencylist_luw_ver1_0.zip',
             'freq': []}}

# BCCWJ データのダウンロードを実行
for k, v in bccwj_info.items():
    
    zip_fname=bccwj_info[k]['zip_fname']
    if not os.path.exists(bccwj_info[k]['zip_fname']):
        url = bccwj_info[k]['url']
        x = requests.get(url, timeout=50)
        #fname=x.headers['Content-disposition'].replace("attachment; filename=\"",'').replace("\"","")
        #zip_fname=bccwj_info[k]['zip_fname']
        if not os.path.exists(zip_fname):
            print(f'{zip_fname} は存在しない。ダウンロードを開始します...', end=" ")
            with open(zip_fname, 'wb') as f:
                f.write(x.content)
            print(' 終了.')

    # ダウンロードしたファイルは，圧縮されている zip ファイルなので，読み込んで解凍
    #zip_fname = os.path.basename(fname)
    tsv_fname = os.path.basename(zip_fname).split('.')[0]+'.tsv'
    if not os.path.exists(tsv_fname):
        with ZipFile(zip_fname, 'r') as z:
            with z.open(tsv_fname) as f:
                file_text = f.readlines()
            # 必要なデータのみ [:8] 取り出し，不要な情報を捨てる
            _freq = [l.decode('utf-8').split('\t')[:8] for l in file_text]
    
        # 頻度情報を freq に格納
        bccwj_info[k]['freq'] = _freq[1:]

for data_name, v in bccwj_info.items():
    print(f'データ名:{colored(data_name, "blue", attrs=["bold"])}', 
          f'総エントリ数:{colored(len(v["freq"]), "blue", attrs=["bold"])}')
````

````python
#1 rank BCCWJ 全体の順位。
#2 lForm 語彙素読み
#3 lemma 語彙素
#4 pos 品詞
#5 subLemma 語彙素細分類
#6 wType 語種
#7 frequency BCCWJ 全体の頻度
#8 pmw BCCWJ 全体での 100 万語当たりの頻度

topn = 5  # 最初の topn エントリだけ表示させる
for x in ['short', 'long']:
    print(f"len(bccwj_info[{x}]['freq']):{colored(len(bccwj_info[x]['freq']),'blue', attrs=['bold'])}")
    for line in bccwj_info[x]['freq'][:topn]:
        print(line)
````

````python
from tqdm.notebook import tqdm

data_name = 'long'  # or 'short'
_data_list = bccwj_info[data_name]['freq']
_data_dict = {}
#_bccwj_ = {}
dups = 0
dup_words = []
_words = []
for lin in tqdm(_data_list[:]):
    lemma = lin[2]
    _words.append(lemma)
    if not lemma in _data_dict:
        _data_dict[lemma] = {'ヨミ': lin[1],
                             'idx': int(lin[0]),
                             '形態素': lin[2],
                             '品詞': lin[3],
                             '語種': lin[5],
                             '頻度': int(lin[6]),
                             'pmw': float(lin[7])}
    else:
        dups +=1 
        dup_words.append(lemma)
        try:
            isinstance(_data_dict[lemma]['dups'], list)
        except:
            _data_dict[lemma]['dups'] = []
        _data_dict[lemma]['dups'].append(lin)

        
_words = sorted(list(set(sorted(_words))))
print(f'len(_words):{colored(len(_words), "blue", attrs=["bold"])}')
for i, wrd in enumerate(_words[:4]):
    print(i, _data_dict[wrd])
print(f'len(dup_words):{colored(len(dup_words), "blue", attrs=["bold"])}')
````

````python
# そうしたら文字種をカウントしてみよう
import numpy as np
import operator
import matplotlib.pyplot as plt
try:
    import japanize_matplotlib
except ImportError:
    !pip install japanize_matplotlib
    import japanize_matplotlib

bccwj_words = _data_dict
print(f'len(bccwj_words):{len(bccwj_words)}\n', 
      f'len(dup_words):{len(dup_words)}\n', 
      f'len(bccwj_words)+len(dup_words):{len(bccwj_words)+len(dup_words)}') # , len(bccwj_short_freq)-1)

chars = {}
for word in tqdm(bccwj_words):
    for ch in word:
        if not ch in chars:
            chars[ch] = 1
        else:
            chars[ch] += 1
print(f'全文字種 len(chars):{colored(len(chars), "blue", attrs=["bold"])}')
topn = 100
print(f'最初の {topn} 文字', list(sorted(chars.keys()))[:topn])
print(f'最後の {topn} 文字', list(sorted(chars.keys()))[-topn:])

chars_count_sorted = sorted(chars.items(), key=operator.itemgetter(1), reverse=True)
N_chr=np.array([v for v in chars.values()]).sum()
````

````python
# そうしたらヨミもカウントしてみよう

_yomi_count = {}
for k, v in bccwj_words.items():
    yomi = v['ヨミ']
    for ch in yomi:
        if not ch in _yomi_count:
            _yomi_count[ch] = 1
        else:
            _yomi_count[ch] += 1
print('ヨミに現わた総文字数:', colored(len(_yomi_count), "blue", attrs=["bold"]))
print('ヨミに現れた全文字:', _yomi_count.keys())

yomi_count_sorted = sorted(_yomi_count.items(), key=operator.itemgetter(1), reverse=True)
N_chr=np.array([v for v in _yomi_count.values()]).sum()
````

````python
# 上の結果から `ヨミ` に現れてはイケナイと思われる文字のリスト
yomi_ng_chars = ['■','ヂ','○','×', '△']

# これらイケナイ文字が含まる単語を計数してみよう
OK_words, NG_words = [], []
for k, v in tqdm(bccwj_words.items()):
    yomi = v['ヨミ']
    wrd = v['形態素']
    idx  = v['idx']
    true_false = True
    for ch in yomi:
        if ch in yomi_ng_chars:
            true_false = False
    if not true_false:
        NG_words.append({'idx':idx, 'wrd':wrd, 'yomi':yomi})
    else:
        OK_words.append({'idx':idx, 'wrd':wrd, 'yomi':yomi})
print(f'len(OK_words):{len(OK_words)}',
      f'len(NG_words):{len(NG_words)}')
````

````python
try:
    import jaconv
except ImportError:
    !pip install jaconv
    import jaconv
    
phones = []
for wrd in tqdm(OK_words):
    yomi = wrd['yomi']
    if 'ヴ' in yomi:
        print(f'juli:{juli}, wrd{wrd}')
        _yomi = "".join(ch if ch != 'ヴ' else 'ブ' for ch in yomi)
        yomi = _yomi
    
    hira = jaconv.kata2hira(yomi)
    juli = jaconv.hiragana2julius(hira).split(' ')
    phones += juli
    
    for p in juli:
        if p in ['a::', 'i::', 'o::', 'o:::', 'u::']:
            print(wrd.values(), juli)
        
phonemes = list(sorted(set(phones)))
print(len(phonemes))
print(phonemes)                
````
